```python
class Max_Page():
    def __init__(self):
        self.Beijing = 110100
        self.Beijing_Region = list(range(110101,110118)) + list(range(110228,110231))

        self.Shanghai = 310100
        self.Shanghai_Region = list(range(310101,310121)) + [310152,310230,310231]

        self.Guangzhou = 440100
        self.Guangzhou_Region = list(range(440103,440108)) + list(range(440111,440117)) + [440188,440189,445400,445500]

        self.Shenzhen = 440300
        self.Shenzhen_Region = list(range(440303,440310))
        
        self.list_bash_page_url= []
        self.list_max_page_text = []
        
    def generate_first_layers(self,city,regions):
        for region in regions:
            url = "http://land.fang.com/market/"+ str(city) +"_" + str(region) + "_1______1_0_1.html"
            self.list_bash_page_url.append(url)
            
    def start(self):
        self.generate_first_layers(self.Beijing,self.Beijing_Region)
        self.generate_first_layers(self.Shanghai,self.Shanghai_Region)
        self.generate_first_layers(self.Guangzhou,self.Guangzhou_Region)
        self.generate_first_layers(self.Shenzhen,self.Shenzhen_Region)
        return self.list_bash_page_url
        
    def scrape(self,url,headers):
        html = requests.get(url,headers = headers)
        html_soup = BeautifulSoup(html.text,'lxml')
        return html_soup

    def scrape_first_layer(self,headers):
        for url in self.list_bash_page_url[0:5]:
            html_soup = self.scrape(url,headers)
            page_max = html_soup.find("div",class_ = "top_page fr").find("span")
            try:
                self.list_max_page_text.append(int(page_max.get_text()[2:4]))
            except Exception as e:
                self.list_max_page_text.append(0)
        return self.list_max_page_text
```

```python
class Formal_Scrape(Max_Page):
    def __init__(self,first_layer_url,max_page_list):
        self.max_page_list = max_page_list
        self.first_layer = first_layer_url
        
        self.second_layer = []
        self.infos_set = []
        
    def generate_second_layer(self):
        for item,max_page in zip(self.first_layer,self.max_page_list):
            if int(max_page) != 0:
                self.second_layer.append(item)
            for j in range(1,max_page + 1):
                item_add = item.replace("1_0_1.html","1_0_" +str(j)+".html")
                self.second_layer.append(item_add)
        print(self.second_layer[0:5])
    
    def scrape_second_layer(self,headers):
        for url in self.second_layer:
            html_soup = self.scrape(url,headers)
            try:
                infos = html_soup.find("dl",class_ = "llist01").find_all("dd")
                print(infos)
                self.infos_set.extend(infos)
            except Exception as e:
                pass
        return self.infos_set
    
    def clean(self,infos):
        TITLE = list(map(lambda x:x.find_all("a")[0].get_text(),infos_set))
        DATE = list(map(lambda x:x.find_all("td")[0].get_text(),infos_set))
        STATUS = list(map(lambda x:x.find_all("td")[1].get_text(),infos_set))
        SQUARE = list(map(lambda x:x.find_all("td")[2].get_text(),infos_set))
        AREA = list(map(lambda x:x.find_all("td")[3].get_text(),infos_set))
        PLAN_SQUARE = list(map(lambda x:x.find_all("td")[4].get_text(),infos_set))
        PLAN_USE = list(map(lambda x:x.find_all("td")[5].get_text(),infos_set))
        table = pd.DataFrame([TITLE,DATE,STATUS,SQUARE,AREA,PLAN_SQUARE,PLAN_USE]).T
        return table
```

```python
if __name__ == "__main__":
    ua = UA()
    headers = ua.UA_Choice()
    mp = Max_Page()
    first_layer_url = mp.start()  
    max_page_list = mp.scrape_first_layer(headers)
    fs = Formal_Scrape(first_layer_url,max_page_list)
    fs.generate_second_layer()
    infos_set = []
    infos_set = fs.scrape_second_layer(headers)
```
