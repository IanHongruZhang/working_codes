```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36 Edge/15.15063'}
```

```python
list_bs4 = []
for i in range(1,13):
    if i < 10:
        url = "http://www.runchina.org.cn/portal.php?mod=calendar&ac=list&year=2017&month=0" + str(i)
        html = requests.get(url,headers = headers)
        html_soup = BeautifulSoup(html.text,'lxml')
        list_bs4.append(html_soup)
    else:
        url = "http://www.runchina.org.cn/portal.php?mod=calendar&ac=list&year=2017&month=" + str(i)
        html = requests.get(url,headers = headers)
        html_soup = BeautifulSoup(html.text,'lxml')
        list_bs4.append(html_soup)
        
print(len(list_bs4))
```

```python
list_name = list(map(lambda x:x.find("ul",class_ = "match-list").find_all("span",class_ = "match-name text-omit"),list_bs4))
```

```python
list_name_individual = []
for month in list_name:
    for item in month:
        item = item.get_text()
        list_name_individual.append(item)
```

```python
import jieba

list_seg_words = []
for item in list_name_individual:
    ls = (list((jieba.cut(item))))
    list_seg_words.extend(ls)
```

```python
from collections import Counter
dict_a = dict(Counter(list_seg_words))

sorted(dict_a.items(),key = lambda x:x[1],reverse = True)
```
