```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
table = pd.read_excel("China_1300_Export.xlsx")
```

```python
table
```

```python
list_2_digits = []
list_4_digits = []
list_6_digits = []
for item in list(table["HTS_Code"]):
    list_2_digits.append(int(str(item)[0:2]))
    list_4_digits.append(int(str(item)[0:4]))
    list_6_digits.append(int(str(item)[0:6]))
```

```python
table_a = pd.DataFrame(list_2_digits) 
table_b = pd.DataFrame(list_4_digits) 
table_c = pd.DataFrame(list_6_digits)
```

```python
table_a.columns = ["2-digits"]
table_b.columns = ["4-digits"]
table_c.columns = ["6-digits"]
```

```python
table_final = table.join([table_a,table_b,table_c])
```

```python
table_final.to_excel("China_1300_Export_Final.xlsx")
```
